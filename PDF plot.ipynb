# Step 2: Plotting PDF from a .csv file

```python
import pandas as pd
import seaborn as sns

#number of bins
n = int(input('number of bins... '))

#locate the file with data
cost = pd.read_csv('data.csv')

#plot function
sns.distplot(cost['Data '], hist=True, kde=True, 
             bins=n, color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4})
```
